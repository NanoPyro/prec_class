Is there a correlation between holiday/vacation seasons and delays?  
We'll try to find that below.

```
df = data.frame(day=as.Date(character()), delay=integer()) # delay varibale is the average value of the delay, which is sum of arr_delay and dep_delay, of that day.

for (i in 1:12)
{
  if ( i %in% c(1,3,5,7,8,10,12) ) # Months that have 31 days
  {
    for (j in 1:31)
    {
      flights %>% filter(month == i & day == j) %>% select(dep_delay, arr_delay) %>% apply(2, mean, na.rm=TRUE) %>% sum() -> tmp_delay

      if (i < 10)
        month = paste0("0", i)
      else
        month = as.character(i)
      if(j < 10)
        day = paste0("0", j)
      else
        day = as.character(j)
    
      date = paste0("2013", "-" ,month, "-", day)
      date = as.Date(date)

      df = rbind(df, data.frame(day=date, delay=tmp_delay))
    }
  }
  else if ( i == 2 ) # February
  {
    for (j in 1:28)
    {
      flights %>% filter(month == i & day == j) %>% select(dep_delay, arr_delay) %>% apply(2, mean, na.rm=TRUE) %>% sum() -> tmp_delay

      month = "02"
      if(j < 10)
        day = paste0("0", j)
      else
        day = as.character(j)
    
      date = paste0("2013", "-" ,month, "-", day)
      date = as.Date(date)

      df = rbind(df, data.frame(day=date, delay=tmp_delay))
    }
  }
  else # Months that have 30 days
  {
    for ( j in 1:30)
    {
      flights %>% filter(month == i & day == j) %>% select(dep_delay, arr_delay) %>% apply(2, mean, na.rm=TRUE) %>% sum() -> tmp_delay

      if (i < 10)
        month = paste0("0", i)
      else
        month = as.character(i)
      if(j < 10)
        day = paste0("0", j)
      else
        day = as.character(j)
    
      date = paste0("2013", "-" ,month, "-", day)
      date = as.Date(date)

      df = rbind(df, data.frame(day=date, delay=tmp_delay))
    }
  }
}
```

```
dim(df) # 365 x 2
df
```

```
ggplot(df, aes(x=day, y=delay)) + geom_line()
```

As we expected, there are huge spikes in summer vacation season and Christmas and holiday season.
